### Conclusion:
From the given image, two objects are extracted as pathches.

The same objects are resized to 128 rows and 64 columns.

The size of each cell is (8,8) and each block contains (2,2) cells. Number of bins is 9.

For the above set of parameters, visualsiation of HOG shows the brighter part on edges of the object and dark pixels where the image has pixels with constant intensity. (patches inside the cup other than the drawing and text)

Length of feature vector is 3780. (7 x 15) x 36 values per block.


---

If cell size is increased from 8x8 to 16x16 or 32x32 it shows broader details as multiple local features are combined into one cell and the finer details are overlooked.

Length of feature vectors:

1. 8x8 - 3780
2. 16x16 - 756
3. 32x32 - 108

It is observed that length of feature vector is reduced as less number of cells will be there in the resized image.


---

Euclidean distance between the feature vectors of two images which are similar, we get a value of 5.63. Euclidean distance is small if the distance between two objects are similar.

---

If we increase the cells per block the visualisation difference is not seen. But when the euclidean distance between the two feature vectors are compared, as the number of cells per block increases the distance between the feature vectors decreases from 5.63 for (2,2), 5.01 for (4,4) and 1.94 for (8,8)


Hence, it can be concluded that HOG can be used to detect the edges and feature vectors can be used to compare two images.
